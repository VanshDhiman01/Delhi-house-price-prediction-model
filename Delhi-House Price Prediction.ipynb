```python
#converting catagorical values to numeric 
new_fields={
    'Transaction':{'New_Property':0,'Resale':1},
    'Type':{'Apartment':0,'Builder_Floor':1},
    'Status':{'Ready_to_move':0,'Almost_ready':1},
    'Furnishing':{'Semi-Furnished':1,'Unfurnished':0,'Furnished':2}
    }
df3=df2.replace(new_fields)
```

```python
dummies=pd.get_dummies(df3['Locality'],dtype=int)
```

```python
df4=df3.drop(['Locality'],axis=1)
```

```python
df5=pd.concat([df4,dummies.drop('Alaknanda',axis=1)],axis=1) 
```

```python
X=df5.drop('Price',axis=1)
```

```python
Y=df5['Price']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
print(X.shape,X_train.shape,X_test.shape)
```
